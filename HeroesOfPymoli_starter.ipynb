```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
# Find the total number of players
total_players = len(purchase_data_df["SN"].unique())
total_players

summary_table_df = pd.DataFrame({"Total Players": [total_players]})

# Display the total number of players
summary_table_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Find the amount of unique item names
unique_items = len(purchase_data_df["Item Name"].unique())
unique_items

# Find the average price per item and format
average_price = "${:,.2f}".format(purchase_data_df["Price"].mean())
average_price

# Find the total number of purchased items
total_purchases = len(purchase_data_df["Purchase ID"])
total_purchases

# Find the total revenue made and format
total_revenue = "${:,.2f}".format(purchase_data_df["Price"].sum())
total_revenue

# Create a summary data frame to hold the results
summary_table_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                "Average Price": [average_price],
                                "Number of Purchases": [total_purchases],
                                "Total Revenue": [total_revenue]})

# Display the summary data frame
summary_table_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Find the percentage and count of Male Players
male_df = purchase_data_df[purchase_data_df["Gender"]=="Male"]
male_count = len(male_df["SN"].unique())
male_percentage = "{:.2%}".format(male_count / total_players)
male_percentage

# Copy & Paste, change to Female
female_df = purchase_data_df[purchase_data_df["Gender"]=="Female"]
female_count = len(female_df["SN"].unique())
female_percentage = "{:.2%}".format(female_count / total_players)
female_percentage

# Copy & Paste, change to Other
other_df = purchase_data_df[purchase_data_df["Gender"]=="Other / Non-Disclosed"]
other_count = len(other_df["SN"].unique())
other_percentage = "{:.2%}".format(other_count / total_players)
other_percentage

# Define the Data Frame of the Summary Table
summary_table_df = pd.DataFrame({"Total Count": [male_count_df, female_count_df, other_count_df],
                                "Percentage of Players": [male_percentage, female_percentage,
                                                         other_percentage]},
                               index = ["Male", "Female", "Other / Non-Disclosed"])

# Display the Summary Table
summary_table_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Find the purchase details of Female Players
female_purchase_count = len(female_df["Purchase ID"].unique())
female_purchase_price = female_df["Price"].sum()
avg_female_purchase_price = "${:.2f}".format(female_purchase_price / female_purchase_count)
female_unique_price = female_df.groupby("SN")["Price"].sum()
female_total_value = "${0:,.2f}".format(female_unique_price.sum())
avg_female_total = "${:.2f}".format(female_unique_price.mean())

# Find the purchase details of Male Players
male_purchase_count = len(male_df["Purchase ID"].unique())
male_purchase_price = male_df["Price"].sum()
avg_male_purchase_price = "${:.2f}".format(male_purchase_price / male_purchase_count)
male_unique_price = male_df.groupby("SN")["Price"].sum()
male_total_value = "${0:,.2f}".format(male_unique_price.sum())
avg_male_total = "${:.2f}".format(male_unique_price.mean())

# Find the purchase details of Other / Non-Disclosed Players
other_purchase_count = len(other_df["Purchase ID"].unique())
other_purchase_price = other_df["Price"].sum()
avg_other_purchase_price = "${:.2f}".format(other_purchase_price / other_purchase_count)
other_unique_price = other_df.groupby("SN")["Price"].sum()
other_total_value = "${0:,.2f}".format(other_unique_price.sum())
avg_other_total = "${:.2f}".format(other_unique_price.mean())


# Create a summary table to hold the results
gender_df = pd.DataFrame({"Gender": ["Female", "Male", "Other / Non-Disclosed"],
                                "Purchase Count": [female_purchase_count, male_purchase_count,
                                                  other_purchase_count],
                                "Average Purchase Price": [avg_female_purchase_price, avg_male_purchase_price,
                                                          avg_other_purchase_price],
                                "Total Purchase Value": [female_total_value, male_total_value,
                                                        other_total_value],
                                 "Avg Total Purchase per Person": [avg_female_total, avg_male_total,
                                                                     avg_other_total]
                                })

# Set Data Frame index as Gender
gender_df = gender_df.set_index("Gender")

# Display the Summary Table
summary_table_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 45]

# Create the groups for the age bins
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data_df["Age Demographics"] = pd.cut(purchase_data_df["Age"], bins, labels=age_groups)

age_group_count = purchase_data_df.groupby("Age Demographics")["SN"].nunique()
age_group_percentage = (age_group_count / total_players).map("{:.2%}".format)

# Create summary dataframe to hold the results
age_demographics_df = pd.DataFrame({"Total Count" : age_group_count, 
                             "Percentage of Players": age_group_percentage}, index = age_groups)

# Display Age Demographics Table
age_demographics_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# retrieve the SN, Purchase ID, and Price columns
top_spenders = purchase_data_df.groupby(['SN', 'Purchase ID', 'Price'])
spenders = top_spenders.size()

# calculate the stats for count, average, and total value
player_purchase_count = spenders.groupby(["SN"]).count()
player_avg_purchase_price = purchase_data_df.groupby(by='SN').mean()['Price']
player_total_purchase_value = purchase_data_df.groupby(by='SN').sum()['Price']

# create a Summary Table
top_spenders_df = pd.DataFrame({"Purchase Count": player_purchase_count,
                               "Average Purchase Price": player_avg_purchase_price,
                               "Total Purchase Value": player_total_purchase_value})
# sort in descending order                                            
sorted_top_spenders = top_spenders_df.sort_values("Total Purchase Value", ascending=False)

# format
sorted_top_spenders["Average Purchase Price"] = sorted_top_spenders["Average Purchase Price"].map("${:.2f}".format)
sorted_top_spenders["Total Purchase Value"] = sorted_top_spenders["Total Purchase Value"].map("${:.2f}".format)

# display a preview of the summary data frame
sorted_top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# retrieve the Item ID, Item Name, and Item Price columns
most_pop = purchase_data_df[['Purchase ID', 'Item Name', 'Price']]

# group by Item Id and Item Name. Perform calculations to obtain purchase count, avg. item price, and total purchase
item_count = purchase_data_df.groupby(by='Item Name')['Price'].count()
item_avg_price = purchase_data_df.groupby(by='Item Name')['Price'].mean().map("${:.3}".format)
item_total_purchase = purchase_data_df.groupby(by='Item Name')['Price'].sum()

# create a summary data frame to hold the results
most_pop_df = pd.DataFrame({"Purchase Count": item_count,
                           "Item Price": item_avg_price,
                           "Total Purchase Value": item_total_purchase})

# sort in descending order                                            
sorted_most_pop_df = most_pop_df.sort_values("Purchase Count", ascending=False)

# format
sorted_most_pop["Total Purchase Value"] = sorted_most_pop["Total Purchase Value"].map("${:.2}".format)

# display a preview of the summary data frame
sorted_most_pop_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# retrieve the Item ID, Item Name, and Item Price columns
most_pop = purchase_data_df[['Purchase ID', 'Item Name', 'Price']]

# group by Item Id and Item Name
item_count = purchase_data_df.groupby(by='Item Name')['Price'].count()
item_avg_price = purchase_data_df.groupby(by='Item Name')['Price'].mean().map("${:.3}".format)
item_total_purchase = purchase_data_df.groupby(by='Item Name')['Price'].sum()

# create a summary data frame to hold the results
most_pop_df = pd.DataFrame({"Purchase Count": item_count,
                           "Item Price": item_avg_price,
                           "Total Purchase Value": item_total_purchase})

# sort in descending order                                            
sorted_most_pop_df = most_pop_df.sort_values("Total Purchase Value", ascending=False)

# format
sorted_most_pop["Total Purchase Value"] = sorted_most_pop["Total Purchase Value"].map("${:.2}".format)

# display a preview of the summary data frame
sorted_most_pop_df.head()
```
